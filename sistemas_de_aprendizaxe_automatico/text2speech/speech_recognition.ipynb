```python
import speech_recognition as sr

audio_file_path = "C:\\Users\\daniel.martinezcarre\\Desktop\\Repositorios\\ia_repositorio\\sistemas_de_aprendizaxe_automatico\\speech2text\\audio_harvard.wav"

recognizer = sr.Recognizer()
with sr.AudioFile(audio_file_path) as source:
    audio_data = recognizer.record(source)
    text = recognizer.recognize_google(audio_data, language = 'en', show_all = False )

print(f"Texto transcrito: {text}")
```

```python
# Generar audio
!pip install -q gtts

from gtts import gTTS
from IPython.display import Audio

tts = gTTS(text, lang='en', tld="com")
tts.save('output.wav')
Audio('output.wav', autoplay=True)
```
