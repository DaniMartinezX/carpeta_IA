# Matplotlib
Biblioteca para la generación de gráficos en dos dimensiones, a partir de datos contenidos en listas o arrays en el lenguaje de programación Python

```python
import numpy as np
import matplotlib.pyplot as plt
```

> El **subplots** nos da la figura (fig) y el axis (ax); este último es el que utilizamos para coger de referencia y dibujar.

```python
fig = plt.figure(figsize=(6, 4))
# ax = fig.add_subplot(1, 1, 1)   # Filas, Columnas, Altura del eje 'Y'
x = [2,3,4,5,6]
y = [4,9,16,25,36]

# El subplots no da la figura y el axis.
fig, ax = plt.subplots()
# Primero obtenemos el puntero del subgrafo y después dibujamos.
ax.plot(x,y)
```

```python
fig, ax = plt.subplots()
x = np.linspace(0, 7, 80) # Empieza en 0, acaba en 7, y que ponga 20 números entre ellos.
y = np.sin(x)
ax.plot(x,y, label= 'Esta es sin', marker= 'o')
y2 = np.cos(x)

ax.plot(x,y2, label= 'Esta es cos', marker= 'o')    # El marker es para cambiar el diseño
ax.set_title("Una prueba de plt")
ax.legend() #Para mostrar la leyenda de los datos puestos con el label

# Si queremos enfocarnos en una parte de la gráfica
ax.set_xlim(0,2)
ax.set_ylim(0.5,1)
ax.annotate("Aquí se cruza", xy=(0.76,0.7), xytext=(1.25,0.8), arrowprops=dict(facecolor="black", shrink=0.05))
ax.plot([0, 0.78], [0.71, 0.71], color="gray", linestyle="--")
ax.plot([0.78, 0.78], [0, 0.71], color="gray", linestyle="--")

# Para dibujar un punto
ax.scatter(0.78, 0.71, s=80, c='red')


ax.text(0.4, 0.72, "(0.78, 0.71)")
```

```python
alfa = 0.7
beta = 10
x = np.linspace(0, 2*np.pi, 1000)
y = np.exp(-alfa*x) * np.sin(beta*x)
fig, ax = plt.subplots()

#fig = plt.figure(figsize=(15,10))   # En PULGADAS
#ax.plot(x,y)

# Para pintar varias gráficas
ax = fig.add_subplot(2,2,1)
ax.plot(x,y)
ax = fig.add_subplot(2,2,2)
ax.plot(2*x,y)
ax = fig.add_subplot(2,2,3)
ax.plot(0.1*x,y)
ax = fig.add_subplot(2,2,4)
ax.plot(0.5*x,y)
```

```python
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(1,1,1)
ax.plot(x,y, label= "función exp(-alfa*x) * np.sin(beta*x)", color= 'green', linewidth=3 )
ax.legend()
ax.grid()   # Para poner las cuadrículas
ax.set_xlabel("Estas son las x")
ax.set_ylabel("Estas son las y")
```
