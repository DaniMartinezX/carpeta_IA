# Descenso del gradiente
Se utiliza en las redes neuronales, es muy importante.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# y = x^2 + 7x - 3

def calcula_y(x):
    y = (x**2) + 7*x - 3
    return y
```

```python
x = np.array([1,2,3,4]).astype(float)
y = calcula_y(x)
print(f"X->{x} Y->{y}")
```

```python
lx = np.linspace(-15, 10, 200)
ly = calcula_y(lx)
plt.plot(lx, ly, color='red')
plt.plot([4, 6], [calcula_y(4), calcula_y(6)], color='b')
# La tangente azul es la llamada derivada
# Según el signo de la derivada nos muestra por donde tenemos que ir
```

```python
print(f'min y = {ly.min()} index = {np.where(ly == ly.min())[0]} x = {lx[np.where(ly == ly.min())[0]]}')
```

```python
plt.plot([4, 6], [calcula_y(4), calcula_y(6)], color='r')
```
