# Descenso del gradiente
Se utiliza en las redes neuronales, es muy importante.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# y = x^2 + 7x - 3

def calcula_y(x):
    y = (x**2) + 7*x - 3
    return y

def derivada_y(x):
    return x*2 + 7
```

```python
x = np.array([1,2,3,4]).astype(float)
y = calcula_y(x)
print(f"X->{x} Y->{y}")
```

### Seg√∫n el signo de la derivada nos muestra por donde tenemos que ir

```python
lx = np.linspace(-15, 10, 200)
ly = calcula_y(lx)
plt.plot(lx, ly, color='red')
plt.plot([4, 6], [calcula_y(4), calcula_y(6)], color='b')
# La tangente azul es la llamada derivada
```

```python
x = -13
y = calcula_y(x)
print(f'x = {x} y = {y}')
lr = 0.1
px = [x]
py = [y]
```

```python
for paso in range(10):    
    x_incrementada = x + 0.1
    y_incrementada = calcula_y(x_incrementada)
    print(f"x_incrementada = {x_incrementada} y_incrementada = {y_incrementada}")
    inc_x = x_incrementada - x
    inc_y = y_incrementada - y
    d = inc_y / inc_x
    print(f"inc_x: {inc_x} inc_y: {inc_y} d: {d}")
    x = x - lr * d
    y = calcula_y(x)
    print(f"x nueva = {x} y nueva= {y_incrementada}")
    px.append(x)
    py.append(y)
    plt.plot(lx, ly, color="r")
    plt.scatter(px, py, color="b")
```

```python
x = -13
y = calcula_y(x)
print(f"x = {x} y = {y}")
lr = 0.7 # leaning_rate
px = [x]
py = [y]

for paso in range(10):
  x_incrementada = x + 0.1
  y_incrementada = calcula_y(x_incrementada)
#  print(f"x_incrementada = {x_incrementada} y_incrementada = {y_incrementada}")
  inc_x = x_incrementada - x
  inc_y = y_incrementada - y
  d = inc_y / inc_x
#  print(f"inc_x: {inc_x} inc_y: {inc_y} d: {d}")
  x = x - lr * d
  y = calcula_y(x)
  print(f"x: {x} y: {y}")
  px.append(x)
  py.append(y)
  plt.plot(lx, ly, color="r")
  plt.scatter(px, py, color="b")
```
