## Para añadir personalmente peso a una imagen en específico

```python
img1 = cv.cvtColor(src1, cv.COLOR_RGB2GRAY)
img2 = cv.cvtColor(src2, cv.COLOR_RGB2GRAY)
img4 = cv.cvtColor(src4, cv.COLOR_RGB2GRAY)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3)

r1 = cv.subtract(img2, img1)
r2 = cv.subtract(img4, img1)
r3 = cv.add(r1, r2)
r3 = cv.addWeighted(r1, 0.5, r2, 0.5)   # Para añadir más peso a una imágen que a otra

ax1.imshow(r1, cmap="gray")
ax2.imshow(r2, cmap="gray")
ax3.imshow(r3, cmap="gray")
```

## Multiplicación

```python
img1 = cv.cvtColor(src1, cv.COLOR_RGB2GRAY)
img2 = cv.cvtColor(src2, cv.COLOR_RGB2GRAY)
img3 = cv.cvtColor(src3, cv.COLOR_RGB2GRAY)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3)

# RESTA
r1 = cv.subtract(img2, img1)
r2 = cv.subtract(img3, img1)
# MULTIPLICACIÓN
r3 = cv.multiply(r1, r2)

ax1.imshow(r1, cmap="gray")
ax2.imshow(r2, cmap="gray")
ax3.imshow(r3, cmap="gray")
```

## División

```python
img1 = cv.cvtColor(src1, cv.COLOR_RGB2GRAY)
img2 = cv.cvtColor(src2, cv.COLOR_RGB2GRAY)
img3 = cv.cvtColor(src3, cv.COLOR_RGB2GRAY)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3)

# DIVISIÓN
d1 = cv.divide(img1, img2)

ax1.imshow(img1, cmap="gray")
ax2.imshow(img2, cmap="gray")
ax3.imshow(d1, cmap="gray")
```

## Binarización
La idea principal de la binarización es crear máscaras.

```python
src1 = cv.imread("C:\\Users\\daniel.martinezcarre\\Desktop\\Repositorios\\ia_repositorio\\Programacion de IA\\tirmestre_2\\imagenes\\test.png"); src1 = cv.cvtColor(src1, cv.COLOR_BGR2RGB)
img1 = cv.cvtColor(src1, cv.COLOR_BGR2GRAY)

umbral = 127    # De la media para abajo negra, de la media para arriba blanco
img1[img1 <= umbral] = 0
img1[img1 > umbral] = 255


plt.imshow(img1, cmap="gray")
```
