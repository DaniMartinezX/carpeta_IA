```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

granos = cv.imread("C:\\Users\\daniel.martinezcarre\\Desktop\\Repositorios\\ia_repositorio\\Programacion de IA\\tirmestre_2\\imagenes\\carretera.jpg")

img = cv.cvtColor(granos, cv.COLOR_BGR2GRAY)
plt.imshow(img, cmap="gray")
```

```python
gray = cv.equalizeHist(img)
img_blur = cv.GaussianBlur(gray, (7,7), 0)
edges = cv.Canny(img_blur, 30, 200)
# plt.imshow(edges, cmap="gray")

# Encontrar lneas
min_votos = 10
min_len = 10 # Mnima logitud de lnea
max_gap = 5 # Mxima separacin entre trozos de la misma lnea
lineas = cv.HoughLinesP(edges, 1, np.pi/180, min_votos, min_len, max_gap) #ndarray
img2 = img.copy()
for linea in lineas:
    # Cogemos cada lnea de la lista de lneas y la dibujamos en la imagen
    x1, y1, x2, y2 = linea[0]
    cv.line(img2, (x1,y1), (x2,y2), (255,0, 0), 2)  # Desde donde empieza hasta donde acaba de ancho 2
plt.imshow(img2)
```
