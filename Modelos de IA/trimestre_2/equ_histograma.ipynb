```python
import cv2 as cv
import numpy as np

from matplotlib import pyplot as plt
import pylab # this allows you to control figure size
pylab.rcParams['figure.figsize'] = (10.0, 8.0) # this controls figure size in the notebook

src = cv.imread("C:\\Users\\daniel.martinezcarre\\Desktop\\Repositorios\\ia_repositorio\\Programacion de IA\\tirmestre_2\\imagenes\\test.png")
src = cv.cvtColor(src, cv.COLOR_BGR2RGB)
plt.imshow(src)
```

> Descompongo y muestro los 3 canales:

```python
fig, (ax) = plt.subplots(1,3)
fig.suptitle("R G B")
ax[0].imshow(src[:,:,0], cmap="gray", vmin=0, vmax=255)
ax[1].imshow(src[:,:,1], cmap="gray", vmin=0, vmax=255)
ax[2].imshow(src[:,:,2], cmap="gray", vmin=0, vmax=255)
```

> Calculo el histograma de cada canal y lo ploteo:

```python
fig, (ax) = plt.subplots(1,3)
fig.suptitle("R G B")
                          #Mcara #Tamao
ax[0].plot(cv.calcHist([src], [0], None, [256], [0,255]), color = "red")
ax[1].plot(cv.calcHist([src], [1], None, [256], [0,255]), color = "green")
ax[2].plot(cv.calcHist([src], [2], None, [256], [0,255]), color = "blue")
```

### Utilidad de ecualizar 
La utilidad de ecualizar una imagen es **mejorar su contraste**.<br>
El contraste es la diferencia entre los tonos ms claros y ms oscuros de una imagen.<br>
El resultado de la ecualizacin maximiza el contraste de una imagen sin perder informacin de tipo estructural, esto es, conservando su entropa 

```python
fig, (ax) = plt.subplots(1,3)
fig.suptitle("R G B Equalizados")
ax[0].imshow(cv.equalizeHist(src[:,:,0]), cmap="gray", vmin=0, vmax=255)
ax[1].imshow(cv.equalizeHist(src[:,:,1]), cmap="gray", vmin=0, vmax=255)
ax[2].imshow(cv.equalizeHist(src[:,:,2]), cmap="gray", vmin=0, vmax=255)
```

```python
merged = cv.merge([cv.equalizeHist(src[:,:,0]),cv.equalizeHist(src[:,:,1]),cv.equalizeHist(src[:,:,2])])
```

```python
fig, axs = plt.subplots(1,3)
fig.suptitle("R G B Merged")

axs[0].plot(cv.calcHist([merged], [0], None, [256], [0,255]), color = "red")
axs[1].plot(cv.calcHist([merged], [1], None, [256], [0,255]), color = "green")
axs[2].plot(cv.calcHist([merged], [2], None, [256], [0,255]), color = "blue")
```
